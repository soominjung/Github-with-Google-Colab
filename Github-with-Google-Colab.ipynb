**1. 런타임에 구글 드라이브 마운트**

```
from google.colab import drive
drive.mount('/content/drive/')
```

***내 드라이브 내에 있는 폴더와 파일 리스트 확인**

폴더 생성 필요 시 구글 드라이브에서 생성

```
! ls drive/'My Drive'
```

**2. git user name과 email을 설정**

```
! git config --global user.email "<email@address>"
! git config --global user.name "<username>"
```

**3. 구글 드라이브 내에 github 저장소를 clone** *(username과  password, repository name 수정 필요)*

이 과정은 처음 한번만 필요 (런타임은 시간이 지나면 사라지지만, 구글 드라이브에 생성된 저장소는 사라지지 않기 때문)

```
! git clone https://<github_username>:<github_password>@github.com/<github_username>/<github_repository_name>.git <path to clone github repository>
```

***저장소가 잘 clone 되었는지 확인**

```
! ls drive/'My Drive'/<path of the repository cloned>
```

**4. Colab에서 파일 생성 혹은 수정 후 add**

(새로운 파일을 생성할 땐 저장소 디렉터리 내에 생성)

```
! git --git-dir=drive/'My Drive'/<path of the repository cloned>/.git/ --work-tree=drive/'My Drive'/<path of the repository cloned> add <created or modified files>
```

**5. add한 항목들을 commit**

```
! git --git-dir=drive/'My Drive'/<path of the repository cloned>/.git/ --work-tree=drive/'My Drive'/<path of the repository cloned>/ commit -m "<commit message>"
```

**6. commit을 github에 push**

```
! git --git-dir=drive/'My Drive'/<path of the repository cloned>/.git/ --work-tree=drive/'My Drive'/<path of the repository cloned>/ push origin master
```
